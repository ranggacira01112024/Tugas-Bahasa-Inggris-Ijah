```python
import pandas as pd
from IPython.display import display

# Create list of sentences and their components
sentences = [
    {
        "sentence": "I was reading a book when she called.",
        "subject": "I",
        "continuous_action": "reading",
        "interrupting_action": "she called"
    },
    {
        "sentence": "They were playing when it rained.",
        "subject": "They",
        "continuous_action": "playing",
        "interrupting_action": "it rained"
    },
    {
        "sentence": "He was cooking dinner when I arrived.",
        "subject": "He",
        "continuous_action": "cooking",
        "interrupting_action": "I arrived"
    },
    {
        "sentence": "The baby was sleeping when the phone rang.",
        "subject": "The baby",
        "continuous_action": "sleeping",
        "interrupting_action": "the phone rang"
    },
    {
        "sentence": "We were studying when the lights went out.",
        "subject": "We",
        "continuous_action": "studying",
        "interrupting_action": "the lights went out"
    },
    {
        "sentence": "She was swimming when thunder started.",
        "subject": "She",
        "continuous_action": "swimming",
        "interrupting_action": "thunder started"
    },
    {
        "sentence": "The children were dancing when I saw them.",
        "subject": "The children",
        "continuous_action": "dancing",
        "interrupting_action": "I saw them"
    },
    {
        "sentence": "He was working when the accident happened.",
        "subject": "He",
        "continuous_action": "working",
        "interrupting_action": "the accident happened"
    },
    {
        "sentence": "They were watching TV when I came.",
        "subject": "They",
        "continuous_action": "watching",
        "interrupting_action": "I came"
    },
    {
        "sentence": "The cat was running when the dog chased it.",
        "subject": "The cat",
        "continuous_action": "running",
        "interrupting_action": "the dog chased it"
    }
]

# Convert to DataFrame
df = pd.DataFrame(sentences)

# Display full sentences
def display_sentences():
    print("Past Continuous Tense Examples:")
    print("-" * 50)
    for i, row in df.iterrows():
        print(f"{i+1}. {row['sentence']}")

# Display analysis
def analyze_sentences():
    display(df[['subject', 'continuous_action', 'interrupting_action']])

# Run examples
display_sentences()
print("\nSentence Analysis:")
analyze_sentences()
```
